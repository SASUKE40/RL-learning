```python
import buffalo_gym  # noqa: F401
import gymnasium as gym
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm

env = gym.make("Buffalo-v0")
seed = 40

epsilons = [0, 0.005, 0.01, 0.1, 0.5]
n_actions = env.action_space.n
print(f"Number of actions: {n_actions}")


experiment_rewards = []


def moving_average(x, window=50):
    return np.convolve(x, np.ones(window) / window, mode="valid")


def epsilon_greedy(q, epsilon):
    env.reset(seed=seed)
    experiment_rewards = []
    for t in tqdm(range(10000)):
        if np.random.rand() < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(q)
        obs, reward, terminated, truncated, info = env.step(action)
        # print(
        #     f"Action taken: {action} Reward received: {reward} Terminated: {terminated} Truncated: {truncated}"
        # )

        counts[action] += 1
        q[action] += (reward - q[action]) / counts[action]
        experiment_rewards.append(reward)
        if terminated or truncated:
            observation, info = env.reset(seed=seed)
            print(f"Environment reset. New observation: {observation}, Info: {info}")
    print(f"Final Q-values: {q}")
    print(f"Final Counts: {counts}")
    print(f"Found best arm: {np.argmax(q)}")
    plt.plot(moving_average(experiment_rewards, window=50), label=f"Epsilon={epsilon}")


for epsilon in epsilons:
    print(f"Running epsilon-greedy with epsilon={epsilon}")
    q = np.zeros(n_actions)
    counts = np.zeros(n_actions)
    epsilon_greedy(q, epsilon)

true_means = env.unwrapped.offset_targets[0]
true_best_arm = np.argmax(true_means)
optimal_reward = np.max(true_means)

print("True means:", true_means)
print("True best arm:", true_best_arm)
print("Optimal reward:", optimal_reward)
plt.axhline(optimal_reward, color="red", linestyle="--", label="Optimal Reward")
plt.xlabel("Time Steps")
plt.ylabel("Rewards")
plt.title("Rewards over Time")
plt.legend()
plt.show()
```
